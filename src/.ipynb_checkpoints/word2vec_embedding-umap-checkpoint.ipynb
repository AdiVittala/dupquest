### Reducing the document vectors (weighted word2vec vectors)

#### read document vectors for train set

```python
data_folder = '/media/siri/78C6823EC681FD1E/minio/data/dq-data/'
```

```python
#run = 'train'
run = 'test'
```

```python
X1_d2v = pickle.load(open(data_folder+'weighted_mean1_'+run+'.p','rb'))
X2_d2v = pickle.load(open(data_folder+'weighted_mean2_'+run+'.p','rb'))
```

```python
red_dims =50
```

```python
X1_d2v[0].shape
```

#### remove nans and build single array for training umap model

```python
def reshape_and_remove_nans(X):
    nan_idx_lst = []
    new_lst = []
    for i,x in enumerate(X):
        if np.isnan(x).any():
            nan_idx_lst.append(i)
        else:
            new_lst.append(x.reshape((1,-1)))
    return nan_idx_lst, new_lst
```

```python
X1_nan_lst, X1_new = reshape_and_remove_nans(X1_d2v)
```

```python
X2_nan_lst, X2_new = reshape_and_remove_nans(X2_d2v)
```

```python
len(X1_nan_lst)
```

```python
len(X1_new)
```

```python
len(X2_nan_lst)
```

```python
len(X2_new)
```

```python
X_d2v = np.concatenate(np.vstack((X1_new, X2_new)))
```

```python
del X1_new, X2_new
```

```python
X_d2v.shape
```

#### train umap model and reduce dimensions

```python
if run == 'train':
    reducer_d2v = umap.UMAP(random_state=42, n_components=red_dims, n_neighbors=50, min_dist=0.5)
else:    
    reducer_d2v = load(data_folder+'umap_reducer50_full.joblib')
```

```python
t0 = time.time()
if run == 'train':
    X_d2v_red = reducer_d2v.fit_transform(X=X_d2v.astype(np.float))
else:    
    X_d2v_red = reducer_d2v.transform(X=X_d2v.astype(np.float))
t1 = time.time() - t0
print(t1)
```

```python
if run == 'train':    
    dump(reducer_d2v, data_folder+'umap_reducer50_full.joblib') 
```

```python
X_d2v_red.shape
```

#### add back the nans

```python
def create_nan_array(r,c):
    arr = np.empty((r,c))
    arr[:] = np.nan
    return arr
```

```python
arr_size = len(X1_d2v)
arr_size
```

```python
X_d2v_new = np.empty((2*arr_size,red_dims))
dest_idx = 0
src_idx = 0
for i in X1_nan_lst:
    print('{} {} {}'.format(i, dest_idx, src_idx))
    X_d2v_new[dest_idx:i] = X_d2v_red[src_idx:src_idx+i-dest_idx]
    X_d2v_new[i] = create_nan_array(1,red_dims)
    src_idx = src_idx+i-dest_idx
    dest_idx = i+1
for j in X2_nan_lst:
    i = j+arr_size
    X_d2v_new[dest_idx:i] = X_d2v_red[src_idx:src_idx+i-dest_idx]
    X_d2v_new[i] = create_nan_array(1,red_dims)
    src_idx = src_idx+i-dest_idx
    dest_idx = i+1
```

```python
X_d2v_new.shape
```

```python
X_d2v_new[32536:32542]
```

```python
def verify_nans(X, X_new):
    nan_idx_lst = []
    new_lst = []
    for i, tup in enumerate(zip(X, X_new)):
        x, x_new = tup
        if np.isnan(x).any() and not (np.isnan(x_new).any()):
            print('error at {}'.format(i))
        elif not(np.isnan(x).any()) and np.isnan(x_new).any():
            print('error at {}'.format(i))
```

```python
verify_nans(X1_d2v, X_d2v_new[:arr_size])
```

```python
verify_nans(X2_d2v, X_d2v_new[arr_size:])
```

#### persist reduced document vectors for train set

```python
pickle.dump(X_d2v_new[:arr_size], open(data_folder+'d2v_1_'+run+'_red.p', 'wb'))
pickle.dump(X_d2v_new[arr_size:], open(data_folder+'d2v_2_'+run+'_red.p', 'wb'))
```
